```python
df = pd.DataFrame({
    "name": ["tharun", "kumar", "Akhil", "Varma", "Kushal"],
    "dept": ["IT", "HR", "IT", "Finance", "HR"],
    "salary": [52000, 48000, 47000, 75000, 50000],
    "city": ["Vizag", "Hyderabad", "Mumbai", "Chennai", "Bangalor"]
})
print(df)
```

```python
print(df)
print(list(df.columns))
print(df.head())
df.info()
df.describe()
```

```python
raw = pd.DataFrame({
    "order_id": ["1001", "1002", "1003"],
    "amount": ["2500", "1800", "3200"]
})

print(raw["amount"].sum())
```

```python
clean = raw.copy()
```

```python
clean["order_id"] = pd.to_numeric(clean["order_id"])
clean["amount"] = pd.to_numeric(clean["amount"])
```

```python
print(clean["amount"].sum())
```
