## 1. Loops

Loops are used to repeat a block of code multiple times.

- `for` loop: used when the number of iterations is known.
- `while` loop: used when repetition continues until a condition becomes false.

```python
# For Loop Example
for i in range(1, 6):
    print('Iteration:', i)

print('\nWhile Loop Example')
count = 1
while count <= 5:
    print('Count:', count)
    count += 1
```

## 2. Functions

Functions allow us to group reusable code.

Benefits:
- Avoid code repetition
- Improve readability
- Make programs easier to maintain

```python
# Function Example
def greet(name):
    return f'Hello, {name}!'

print(greet('Alif'))

# Function with Return Value
def square(number):
    return number ** 2

print('Square of 5 =', square(5))
```

## 3. List Comprehensions

List comprehensions provide a compact way to create lists.

General Syntax:
`[expression for item in iterable if condition]`

They are often shorter and more readable than traditional loops.

```python
# Traditional Loop
squares = []
for x in range(1, 6):
    squares.append(x ** 2)

print('Using loop:', squares)

# List Comprehension
squares_comp = [x ** 2 for x in range(1, 6)]
print('Using list comprehension:', squares_comp)

# With condition
even_numbers = [x for x in range(1, 11) if x % 2 == 0]
print('Even numbers:', even_numbers)
```

## 4. File Input/Output (File I/O)

File I/O allows programs to read data from files and write data to files.

Common Modes:
- `r` : Read
- `w` : Write (overwrites existing content)
- `a` : Append
- `r+` : Read and Write

```python
# Writing to a file
with open('sample.txt', 'w') as file:
    file.write('Python File I/O Example\n')
    file.write('Learning Python is fun!')

print('File created successfully.')

# Reading from a file
with open('sample.txt', 'r') as file:
    content = file.read()

print('\nFile Content:')
print(content)
```
