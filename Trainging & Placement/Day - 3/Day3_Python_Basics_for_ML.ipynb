## 6. Loops

### For Loop — Example 1

```python
for i in range(5):
    print(i)
```

## 8. Functions

## 9. List Comprehension
List comprehension is a shorter, cleaner way to create a new list from an existing list or iterable in Python. \
It lets you replace a traditional multi-line for loop with a single line of code

1. The Traditional for Loop Way(4 line)

```python
numbers = [1, 2, 3, 4]
squares = []

for x in numbers:
    squares.append(x * x)

print(squares)  # Output: [1, 4, 9, 16]
```

2. The List Comprehension Way (1 line)

```python
numbers = [1, 2, 3, 4]

squares = [x * x for x in numbers]

print(squares)  # Output: [1, 4, 9, 16]
```

**Example 1 — Squares**

```python
squares = [x**2 for x in range(5)]
print(squares)
```

**Example 2 — With a filter condition**

```python
evens = [x for x in range(20) if x % 2 == 0]
print(evens)
```

**Example 3 — Transform a list of strings**

```python
words = ["python", "ml", "data"]
upper = [w.upper() for w in words]
print(upper)
```

**Example 4 — If-else inside comprehension**

```python
nums = [1, 2, 3, 4, 5]
labels = ["even" if n % 2 == 0 else "odd" for n in nums]
print(labels)
```

## 10. Dictionaries

Store data as key-value pairs.

A dictionary in Python is a built-in data structure used to store data in key-value pairs. \
Unlike lists, which look up items by their numerical position (index), dictionaries look up items using a unique label called a key. Think of it like a real-world phone book, where you look up a person's name (the key) to find their phone number (the value).

**Example 1 — Create and access**

```python
student = {"name": "Ali", "age": 22, "marks": 90}
print(student["name"])
print(student["marks"])
```

**Example 2 — Add, update, delete**

```python
student = {"name": "Ali", "age": 22}
student["city"] = "Delhi"   # add
student["age"] = 23          # update
del student["city"]          # delete
print(student)
```

**Example 3 — Loop through a dictionary**

```python
marks = {"Math": 90, "Physics": 85, "Chemistry": 88}
for subject, score in marks.items():
    print(subject, "->", score)
```

**Example 4 — Useful methods**

```python
d = {"a": 1, "b": 2, "c": 3}
print(list(d.keys()))
print(list(d.values()))
print(d.get("z", "not found"))  # safe access
```

## 11. Tuples

Immutable collection.

A tuple in Python is a built-in data structure used to store an ordered collection of items in a single variable.

Tuples are almost identical to lists, with one critical difference: they are immutable, meaning they cannot be changed, altered, or rearranged once they are created.

**Example 1 — Create and access**

```python
coordinates = (10, 20)
print(coordinates[0])
print(coordinates[1])
```

**Example 2 — Tuple unpacking**

```python
point = (3, 4, 5)
x, y, z = point
print(x, y, z)
```

**Example 3 — Tuples in a list (ML-style records)**

```python
records = [("Ali", 90), ("Sara", 85), ("John", 78)]
for name, score in records:
    print(name, score)
```

**Example 4 — Why immutability matters**

```python
t = (1, 2, 3)
try:
    t[0] = 99
except TypeError as e:
    print("Error:", e)
```

## 13. File Handling

Read/write data files for ML projects.

**Example 1 — Write to a file**

```python
with open("data.txt", "w") as f:
    f.write("Hello from Python\n")
print("Written.")
```

**Example 2 — Read a whole file**

```python
with open("data.txt", "r") as f:
    content = f.read()
print(content)
```

**Example 3 — Write multiple lines and read line by line**

```python
with open("lines.txt", "w") as f:
    for i in range(1, 4):
        f.write(f"Line {i}\n")

with open("lines.txt", "r") as f:
    for line in f:
        print(line.strip())
```

**Example 4 — Append mode**

```python
with open("lines.txt", "a") as f:
    f.write("Added later\n")

with open("lines.txt") as f:
    print(f.read())
```

## 14. Exception Handling

**Example 1 — Catch one error**

```python
try:
    result = 10 / 0
except ZeroDivisionError:
    print("Cannot divide by zero")
```

**Example 2 — Catch multiple errors**

```python
try:
    n = int("abc")
except ValueError:
    print("Not a valid number")
except ZeroDivisionError:
    print("Cannot divide by zero")
```

**Example 3 — try / except / else / finally**

```python
try:
    x = int("10")
except ValueError:
    print("Failed")
else:
    print("Success, value =", x)
finally:
    print("This always runs")
```

**Example 4 — Raising your own error**

```python
def sqrt_pos(n):
    if n < 0:
        raise ValueError("n must be non-negative")
    return n ** 0.5

print(sqrt_pos(16))
try:
    sqrt_pos(-4)
except ValueError as e:
    print("Caught:", e)
```
