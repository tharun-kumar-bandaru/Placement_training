```python
import numpy as np
from sklearn.model_selection import train_test_split
```

```python
## 90 students passed (0), 10 student failed (1)
y = np.array ([0] * 90 + [1] * 10)
```

```python
## without stratification
y_train, y_test = train_test_split(y, test_size = 0.2, random_state = 42)
```

```python
print("Without Stratification")
print("Train failure percentage:", round(y_train.mean() * 100, 2), "%")
print("Test failure percentage:", round(y_test.mean() * 100, 2), "%")
```

```python
#with stratification
y_train, y_test = train_test_split(y, test_size = 0.2, stratify = y, random_state = 42) 
```

```python
print("With Stratification")
print("Train failure percentage:", round(y_train.mean() * 100, 2), "%")
print("Test failure percentage:", round(y_test.mean() * 100, 2), "%")
```
