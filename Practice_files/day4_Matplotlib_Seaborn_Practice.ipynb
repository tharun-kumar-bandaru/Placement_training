## Setup

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
sns.set_theme(style="whitegrid")

tips    = sns.load_dataset("tips")
iris    = sns.load_dataset("iris")
titanic = sns.load_dataset("titanic")
```

### Exercise 1 — Line Plot
Create a line plot of `y = x**2` for `x` from -10 to 10. Add title, xlabel, ylabel.

```python
# Your code here
# Create x values from -10 to 10
x = np.arange(-10, 11)
y = x**2

# Create the plot
plt.figure(figsize=(8, 5))
plt.plot(x, y, marker='o', color='blue')
plt.title("Line Plot of y = x²")
plt.xlabel("x")
plt.ylabel("y = x²")
plt.grid(True)
plt.show()
```

### Exercise 2 — Scatter Plot
Using the `iris` dataset, plot `sepal_length` vs `sepal_width` colored by `species` using `sns.scatterplot`.

```python
# Your code here
plt.figure(figsize=(8, 6))

sns.scatterplot(
    data=iris,
    x="sepal_length",
    y="sepal_width",
    hue="species",
    s=80
)

plt.title("Sepal Length vs Sepal Width by Species")
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Sepal Width (cm)")
plt.legend(title="Species")
plt.show()
```

### Exercise 3 — Histogram + KDE
Plot the distribution of `total_bill` from `tips` with a KDE overlay.

```python
# Your code here
plt.figure(figsize=(8, 6))

sns.histplot(
    data=tips,
    x="total_bill",
    kde=True,
    bins=20,
    color="skyblue"
)

plt.title("Distribution of Total Bill")
plt.xlabel("Total Bill ($)")
plt.ylabel("Frequency")
plt.show()
```

### Exercise 4 — Boxplot
Make a boxplot of `total_bill` grouped by `day` in the `tips` dataset. Which day has the most outliers?

```python
# Your code here
plt.figure(figsize=(8, 6))

sns.boxplot(
    data=tips,
    x="day",
    y="total_bill",
    palette="Set2"
)

plt.title("Total Bill by Day")
plt.xlabel("Day")
plt.ylabel("Total Bill ($)")
plt.show()
```

### Exercise 5 — Violin Plot
Plot a violin plot of `age` by `class` in the `titanic` dataset, split by `sex`.

```python
# Your code here
plt.figure(figsize=(10, 6))

sns.violinplot(
    data=titanic,
    x="class",
    y="age",
    hue="sex",
    split=True,
    palette="Set2"
)

plt.title("Age Distribution by Passenger Class and Sex")
plt.xlabel("Passenger Class")
plt.ylabel("Age")
plt.legend(title="Sex")
plt.show()
```

### Exercise 6 — Countplot
Show the count of survivors vs non-survivors in `titanic`, broken down by `sex` (use `hue`).

```python
# Your code here
plt.figure(figsize=(8, 6))

sns.countplot(
    data=titanic,
    x="survived",
    hue="sex",
    palette="Set2"
)

plt.title("Survivors vs Non-Survivors by Sex")
plt.xlabel("Survived")
plt.ylabel("Count")

# Replace 0 and 1 with readable labels
plt.xticks([0, 1], ["No", "Yes"])

plt.legend(title="Sex")
plt.show()
```

### Exercise 7 — Bar Plot
Plot the mean `tip` per `day` from `tips`, with `sex` as hue.

```python
# Your code here
plt.figure(figsize=(8, 6))

sns.barplot(
    data=tips,
    x="day",
    y="tip",
    hue="sex",
    palette="Set2"
)

plt.title("Mean Tip per Day by Sex")
plt.xlabel("Day")
plt.ylabel("Mean Tip ($)")
plt.legend(title="Sex")

plt.show()
```

### Exercise 8 — Pairplot
Generate a `pairplot` of `iris` colored by species. Identify which feature pair best separates the classes.

```python
# Your code here
sns.pairplot(
    data=iris,
    hue="species",
    palette="Set2",
    diag_kind="hist"
)

plt.suptitle("Pairplot of Iris Dataset", y=1.02)
plt.show()
```

### Exercise 9 — Heatmap
Compute the correlation matrix of numeric columns in `titanic` and visualize it with an annotated heatmap.

```python
# Your code here
plt.figure(figsize=(10, 8))

# Select only numeric columns
corr_matrix = titanic.select_dtypes(include="number").corr()

# Plot the heatmap
sns.heatmap(
    corr_matrix,
    annot=True,
    cmap="coolwarm",
    fmt=".2f",
    linewidths=0.5
)

plt.title("Correlation Matrix of Titanic Numeric Features")
plt.show()
```

### Exercise 10 — Subplots
In a single figure with 1 row and 2 columns, show: (left) histogram of `total_bill`, (right) histogram of `tip`.

```python
# Your code here
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# Left subplot: Total Bill
sns.histplot(
    data=tips,
    x="total_bill",
    bins=20,
    kde=True,
    color="skyblue",
    ax=axes[0]
)
axes[0].set_title("Distribution of Total Bill")
axes[0].set_xlabel("Total Bill ($)")
axes[0].set_ylabel("Frequency")

# Right subplot: Tip
sns.histplot(
    data=tips,
    x="tip",
    bins=20,
    kde=True,
    color="salmon",
    ax=axes[1]
)
axes[1].set_title("Distribution of Tip")
axes[1].set_xlabel("Tip ($)")
axes[1].set_ylabel("Frequency")

plt.tight_layout()
plt.show()
```
