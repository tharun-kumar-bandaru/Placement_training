```python
df["Sleep_Hours"] = df["Sleep_Hours"].fillna(df["Sleep_Hours"].mean())
```

```python
df.isnull().sum()
```

```python
df["Study_Hours"] = df["Study_Hours"].fillna(df["Study_Hours"].mean())
df["Social_Media_Hours"] = df["Social_Media_Hours"].fillna(df["Social_Media_Hours"].mean())
df["Attendance"] = df["Attendance"].fillna(df["Attendance"].mean())
df["Exam_Pressure"] = df["Exam_Pressure"].fillna(df["Exam_Pressure"].mean())
df["Family_Support"] = df["Family_Support"].fillna(df["Family_Support"].mean())
df["Month"] = df["Month"].fillna(df["Month"].mean())
```

```python
df.isnull().sum()
```

```python
df["Student_Type"] = df["Student_Type"].fillna(df["Student_Type"].mode()[0])
```

```python
## Formula to remove outliers
Q1 = df["Sleep_Hours"].quantile(0.25) 
Q3 = df["Sleep_Hours"].quantile(0.75)

IQR = Q3 - Q1

lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR
```

```python
outliers = df[(df["Sleep_Hours"] < lower) | (df["Sleep_Hours"] > upper)] ## to see outliers
```

```python
outliers
```

```python
df = df[(df["Sleep_Hours"] >= lower) & (df["Sleep_Hours"] <= upper)]
```

```python
sns.boxplot(df["Sleep_Hours"])
```
